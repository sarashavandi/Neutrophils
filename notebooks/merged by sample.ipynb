```python
# importing the necessary libraries:
import warnings
warnings.filterwarnings("ignore")
import anndata as ad
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import leidenalg
import seaborn as sns
import gseapy as gp
import math
```

```python
neu = sc.read("../data/merged_neu_with_blob.h5ad")
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
sc.pl.umap(
    neu,
    color="sample",
    palette={
        "Tumour": "#6666FF",
        "Wound": "#FF00CC",
    },
    ax=ax,
    #na_color=fallback,
    na_in_legend=True,
    legend_fontsize=16,
    legend_fontoutline=2,
    add_outline=False,
    size=5,
    show=False,
)

ax.set(xlabel="", ylabel="")
ax.title.set_visible(False)
ax.set_axis_off()
ax.set_aspect(1.0)
fig.tight_layout()
plt.show()
```
